```python
data = boston.data[:, np.newaxis, 0]
```

```python
data.shape
```

```python
train_x = data[:450]
```

```python
test_x = data[450:]
```

```python
train_y = boston.target[:450]
```

```python
test_y= boston.target[450:]
```

```python
regress = linear_model.LinearRegression()
```

```python
regress.fit(train_x,train_y)
```

```python
regress.predict(test_x)
```

```python
regress.coef_
```

```python
plt.scatter(regress.predict(test_x),test_y)
```

```python
plt.scatter(test_x,test_y)
```

```python
plt.scatter(test_x,test_y,color='brown')
plt.plot(test_x,regress.predict(test_x),color='green')
```

```python
mean_squared_error=np.mean((regress.predict(test_x)-test_y)**2)
```

```python
mean_squared_error
```

```python
from sklearn.model_selection import cross_val_predict
```

```python
predicted = cross_val_predict(linear_model.LinearRegression(), boston.data, boston.target, cv=5)
```
