```python
plt.plot(arr,np.sin(arr),color='m',linewidth=2)
plt.ylim(-1.5,1.5)
plt.xlim(-3.5,3.5)
plt.show()
```

```python
y_sin=np.sin(arr)
plt.plot(arr,y_sin,color='m',linewidth=2)
plt.ylim(y_sin.min()*1.1,y_sin.max()*1.1)
plt.xlim(arr.min()*1.1,arr.max()*1.1)
plt.show()
```

```python
y_sin=np.sin(arr)
plt.plot(arr,y_sin,color='m',linewidth=2)

plt.ylim(y_sin.min()*1.1,y_sin.max()*1.1)
plt.xlim(arr.min()*1.1,arr.max()*1.1)

plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi])
plt.show()
```

```python
y_sin=np.sin(arr)
plt.plot(arr,y_sin,color='m',linewidth=2)

plt.ylim(y_sin.min()*1.1,y_sin.max()*1.1)
plt.xlim(arr.min()*1.1,arr.max()*1.1)

plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$','0',r'$\pi/2$',r'$\pi/2$'])
plt.show()
```

```python
y_sin=np.sin(arr)
plt.plot(arr,y_sin,color='m',linewidth=2)

plt.ylim(y_sin.min()*1.1,y_sin.max()*1.1)
plt.xlim(arr.min()*1.1,arr.max()*1.1)

plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$','0',r'$\pi/2$',r'$\pi/2$'])
plt.yticks([-1,0,1])
plt.show()
```

```python
y_sin=np.sin(arr)
plt.plot(arr,y_sin,color='m',linewidth=2)

plt.ylim(y_sin.min()*1.1,y_sin.max()*1.1)
plt.xlim(arr.min()*1.1,arr.max()*1.1)

plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$','0',r'$\pi/2$',r'$\pi/2$'])
plt.yticks([-1,0,1])
plt.title('Sin(x)')
plt.legend(['Sin(x)'])
```

```python
np.info(plt.legend)
```

```python
y_sin=np.sin(arr)
plt.plot(arr,y_sin,color='m',linewidth=2,label='sin(x)')

plt.ylim(y_sin.min()*1.1,y_sin.max()*1.1)
plt.xlim(arr.min()*1.1,arr.max()*1.1)

plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$','0',r'$\pi/2$',r'$\pi/2$'])
plt.yticks([-1,0,1])
plt.title('Sin(x)')
plt.legend()
#plt.legend(['Sin(x)'])
```

```python
y_sin=np.sin(arr)
plt.plot(arr,y_sin,color='m',linewidth=2,label='sin(x)')

plt.ylim(y_sin.min()*1.1,y_sin.max()*1.1)
plt.xlim(arr.min()*1.1,arr.max()*1.1)

plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$','0',r'$\pi/2$',r'$\pi/2$'])
plt.yticks([-1,0,1])
plt.title('Sin(x)')
plt.legend(loc='lower right',frameon=False,fontsize='large')
#plt.legend(['Sin(x)'])
```

```python
y_sin=np.sin(arr)
y_cos=np.cos(arr)
plt.plot(arr,y_sin,color='m',linewidth=2,label='sin(x)')
plt.plot(arr,y_cos,color='c',linewidth=2,label='cos(x)')

plt.ylim(y_sin.min()*1.1,y_sin.max()*1.1)
plt.xlim(arr.min()*1.1,arr.max()*1.1)

plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$','0',r'$\pi/2$',r'$\pi/2$'])
plt.yticks([-1,0,1])
#plt.title('Sin(x)')
plt.legend(loc='lower right',frameon=False,fontsize='large')
```
